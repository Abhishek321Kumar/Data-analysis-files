```python
df=pd.read_csv(r'C:\Users\Abhishek\Desktop\Copywriting\Futurelytic_course_practice\Data analytics with Python\data\medical_examination.csv')
df
```

```python
df.info()
```

<b> Adding an overweight column to the data. To determine if a person is overweight, calculate their BMI by dividing their weight in kilograms by the square of their height in meters. If that value is > 25 then the person is overweight. Use the value 0 for NOT overweight and the value 1 for overweight.</b>

```python
df['overweight']=(df['weight'] / ((df['height']/100 )** 2)) > 25
df['overweight']=df['overweight'].astype(int)
df
```

<b>Normalizing data by making 0 always good and 1 always bad. If the value of cholesterol or gluc is 1, set the value to 0. If the value is more than 1, set the value to 1.</b>

```python
df['cholesterol'] = np.where((df['cholesterol'] > 1) | (df['cholesterol']<1) , 1, 0)
df['gluc'] = np.where((df['gluc'] > 1) | (df['gluc']< 1), 1, 0)
df
```

<b>Creating a DataFrame for the categorical plot using pd.melt with values from cholesterol, gluc, smoke, alco, active, and overweight in the df_cat variable.</b>

```python
df.head(20)
```

```python
df_cat =  pd.melt(df, 
                     id_vars=['cardio'], 
                     value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
```

<b>Grouping and reformatting the data in df_cat to split it by cardio, showing the counts of each feature. Renaming one of the columns for the catplot to work correctly.</b>

```python
df_cat= df_cat.groupby(['cardio', 'variable', 'value'], as_index=False).size()

df_cat.rename(columns={'size': 'count'}, inplace=True)

df_cat
```

<b>Converting the data into long format and creating a chart that shows the value counts of the categorical features</b>

```python
plot = sns.catplot(
        x='variable', 
        y='count', 
        hue='value', 
        col='cardio', 
        data=df_cat, 
        kind='bar',
        height=5, 
        aspect=1.2
    )
```

<b>Getting the figure for the output and storing it in the fig variable.</b>

```python
fig = plot.fig
fig.savefig('catplot.png')
```

<b>Cleaning the data in the df_heat variable by filtering out the following patient segments that represent incorrect data:
<ul>
<li>diastolic pressure is higher than systolic (Keeping the correct data with (df['ap_lo'] <= df['ap_hi']))</li>
<li>height is less than the 2.5th percentile (Keeping the correct data with (df['height'] >= df['height'].quantile(0.025)))</li>
<li>height is more than the 97.5th percentile</li>
<li>weight is less than the 2.5th percentile</li>
<li>weight is more than the 97.5th percentile </li></ul></b>

```python
df_heat = df[(df['ap_lo']<=df['ap_hi']) & 
             (df['height'] >= df['height'].quantile(0.025)) & 
             (df['height'] <= df['height'].quantile(0.975)) &
              (df['weight'] >= df['weight'].quantile(0.025)) & 
              (df['weight'] <= df['weight'].quantile(0.975)) ]


df_heat
```

<b>Calculating the correlation matrix and storing it in the corr variable.</b>

```python
corr = df_heat.corr()
corr
```

<b>Generating a mask for the upper triangle and storing it in the mask variable. Setting up the matplotlib figure and plotting the correlation matrix</b>

```python
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(10, 8))

# Draw the heatmap
sns.heatmap(
    corr,
    mask=mask,             # Apply the mask      
    annot=True,            # Annotate cells with correlation values
    fmt='.1f',             # Format for annotations
    linewidths=0.5, 
    vmin=-0.1,
    vmax=0.25 
      
   
)


# Set the title
ax.set_title('Correlation Matrix Heatmap')
```
